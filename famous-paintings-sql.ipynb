### <font color=blue>1. Fetch all the paintings which are not displayed at any museums

```python
%%sql
SELECT *
FROM work
WHERE museum_id IS NULL
LIMIT 10;
```

### <font color=blue>2. Are there any museums without any paintings?

```python
%%sql
SELECT *
FROM museum AS m
LEFT JOIN work AS w
    ON m.museum_id = w.museum_id
    WHERE w.museum_id IS NULL;
```

### <font color=blue>3. How many paintings have an asking price which more than their regular price?

```python
%%sql
SELECT
    COUNT(*) AS nr_of_paintings
FROM product_size
WHERE sale_price > regular_price;
```

### <font color=blue>4. Identify the paintings whose asking price is less than 50% of its regular price

```python
%%sql
SELECT *
FROM product_size
WHERE sale_price < (0.50 * regular_price)
LIMIT 20;
```

### <font color=blue>5. Which canvas size costs the most?

```python
%%sql
SELECT 
  DISTINCT(size_id),
  width,
  height,
  label
FROM (
  SELECT
    c.*,
    p.regular_price
  FROM canvas_size AS c
    INNER JOIN product_size AS p
    ON c.size_id = p.size_id
  WHERE regular_price = (SELECT MAX(regular_price) FROM product_size)) AS max_price
```

### <font color=blue>6. Fetch the top 10 most famous painting subject

```python
%%sql
SELECT 
  subject,
  COUNT(*) AS nr_of_paintings
FROM subject
GROUP BY 1
ORDER BY nr_of_paintings DESC
LIMIT 10;
```

### <font color=blue>7. Identify the museums which are open on both Sunday and Monday.
     Display museum name and city

```python
%%sql
SELECT
  mh.museum_id,
  m.name,
  m.city
FROM museum_hrs AS mh
JOIN museum AS m 
  ON mh.museum_id = m.museum_id
WHERE day = 'Sunday' 
  AND mh.museum_id IN (SELECT museum_id FROM museum_hrs WHERE day = 'Monday');
```

### <font color=blue>8. How many museums are open every single day?

```python
%%sql
SELECT 
  COUNT(*) AS everyday_museums
FROM(
  SELECT *,
    ROW_NUMBER() OVER(PARTITION BY museum_id) AS rn
  FROM museum_hrs) AS museum_hrs_days
  WHERE rn = 7;
```

```python
%%sql
SELECT 
  COUNT(*) AS everyday_museums
FROM (
  SELECT
    museum_id,
    COUNT(DISTINCT day) AS total_days_opened
  FROM museum_hrs
  GROUP BY 1
  HAVING total_days_opened = 7
);
```

### <font color=blue>9. Which are the top 5 most popular museums?
     Popularity is defined based on most nr. of paintings in a museum

```python
%%sql
SELECT  *
FROM(
  SELECT 
    m.name AS museum_name,
    COUNT(w.work_id) AS nr_of_paintings
  FROM museum AS m
  JOIN work AS w
    ON m.museum_id = w.museum_id
  GROUP BY 1) AS museum_painting_counts
ORDER BY nr_of_paintings DESC
LIMIT 5;
```

### <font color=blue>10. Who are the top 5 most popular artists?
     Popularity is based on most nr. of paintings done by an artist

```python
%%sql
SELECT *
FROM (
  SELECT 
    a.full_name AS artist_name,
    COUNT(w.work_id) AS nr_of_paintings
  FROM artist AS a
  JOIN work AS w
    ON a.artist_id = w.artist_id
  GROUP BY 1)AS artist_painting_counts
ORDER BY nr_of_paintings DESC
LIMIT 5;
```

### <font color=blue>11. Display the 3 least popular canvas sizes

```python
%%sql
SELECT *
FROM (
  SELECT
    c.label as canvas_label,
    COUNT(w.work_id) AS nr_of_paintings
  FROM canvas_size AS c
  JOIN product_size AS p
    ON c.size_id = p.size_id
  JOIN work AS w
    ON p.work_id = w.work_id
  GROUP BY 1
)AS canvas_label_painting_counts
ORDER BY nr_of_paintings ASC
LIMIT 3;
```

### <font color=blue>12. Which museum is open for the longest during a day?
     Display museum name, city and hours open and which day

```python
%%sql
SELECT 
    m.name,
    m.city,
    mh.day,
    (mh.open - mh.close)  AS hrs_open
FROM museum_hrs AS mh
JOIN museum AS m
    ON mh.museum_id = m.museum_id
ORDER BY (mh.open - mh.close) DESC
LIMIT 1;
```

### <font color=blue>13. Which museum has the most painting style?

```python
%%sql
SELECT 
    m.name,
    COUNT(w.style) AS nr_of_painting_styles
FROM work AS w
JOIN museum AS m
    ON w.museum_id = m.museum_id
GROUP BY 1
ORDER BY nr_of_painting_styles DESC
LIMIT 1;
```

### <font color=blue>14. Identify the top 5 artists whose paintings are displayed in multiple countries

```python
%%sql
SELECT 
  a.full_name AS artist_name,
  COUNT(*) AS nr_of_countries_displayed
FROM museum AS m
JOIN work AS w
  ON m.museum_id = w.museum_id
JOIN artist AS a
  ON w.artist_id = a.artist_id
GROUP BY 1
HAVING nr_of_countries_displayed > 2
ORDER BY COUNT(*) DESC
LIMIT 5;
```

### <font color=blue>15. Which country has the 5th highest nr. of paintings?

```python
%%sql
SELECT 
  m.country,
  COUNT(*) AS nr_of_paintings
FROM museum AS m
JOIN work AS w
  ON m.museum_id = w.museum_id
GROUP BY 1
ORDER BY COUNT(*) DESC
LIMIT 1
OFFSET 4;
```

### <font color=blue>16. Which are the 3 most popular and 3 least popular painting styles?

```python
%%sql
SELECT style, nr_of_paintings, remarks
FROM (
    SELECT 
        style,
        COUNT(*) AS nr_of_paintings,
        'Most Popular' AS remarks
    FROM work
    GROUP BY style
    ORDER BY COUNT(*) DESC
    LIMIT 3
) AS most_popular
UNION ALL
SELECT style, nr_of_paintings, remarks
FROM (
    SELECT 
        style,
        COUNT(*) AS nr_of_paintings,
        'Least Popular' AS remarks
    FROM work
    GROUP BY style
    ORDER BY COUNT(*) ASC
    LIMIT 3
) AS least_popular;
```
